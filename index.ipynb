### 3.Develop a classification model with high accuracy to predict customer churn for SyriaTel.

### Baseline Model:

Train a simple baseline model to establish a performance baseline for comparison.
Use a common algorithm such as Logistic Regression.

```python
# Drop irrelevant columns, including 'phone number'
X = df.drop(['churn', 'phone number'], axis=1)
y = df['churn']

# One-Hot Encoding for Categorical Variables
X_encoded = pd.get_dummies(X, columns=['state', 'area code', 'international plan', 'voice mail plan'], drop_first=True)

# Train-Test Split for the Encoded Data
X_train_encoded, X_test_encoded, y_train_encoded, y_test_encoded = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Baseline Logistic Regression Model
baseline_model = LogisticRegression(random_state=42)
baseline_model.fit(X_train_encoded, y_train_encoded)

# Predictions
y_pred_baseline = baseline_model.predict(X_test_encoded)

# Evaluate Baseline Model
accuracy_baseline = accuracy_score(y_test_encoded, y_pred_baseline)
classification_report_baseline = classification_report(y_test_encoded, y_pred_baseline)
confusion_matrix_baseline = confusion_matrix(y_test_encoded, y_pred_baseline)

# Display Results
print(f"Baseline Model Accuracy: {accuracy_baseline:.4f}")
print("\nClassification Report:")
print(classification_report_baseline)
print("\nConfusion Matrix:")
print(confusion_matrix_baseline)
```

Interpretation:

Accuracy: The baseline logistic regression model achieved an accuracy of 84.71%, meaning it correctly predicted the churn labels for 84.71% of the instances in the test set.

Classification Report:

Precision: Precision is the ratio of correctly predicted positive observations to the total predicted positives. In this case, precision for the "True" class is 44%, indicating that among the instances predicted as "True," only 44% were actually true positives.

Recall (Sensitivity): Recall is the ratio of correctly predicted positive observations to the all observations in the actual class. Recall for the "True" class is 4%, indicating that only 4% of the actual "True" instances were correctly predicted.

F1-Score: The F1-Score is the weighted average of precision and recall. It is a good way to show a balance between precision and recall. F1-Score for the "True" class is 7%.

The confusion matrix shows the counts of true positive, true negative, false positive, and false negative predictions. In this case, there are 561 true negatives, 5 false positives, 97 false negatives, and 4 true positives.

```python
# Predict probabilities for the positive class
y_prob_baseline = baseline_model.predict_proba(X_test_encoded)[:, 1]

# Calculate AUC-ROC score
roc_auc_baseline = roc_auc_score(y_test_encoded, y_prob_baseline)

# Generate ROC curve data
fpr_baseline, tpr_baseline, _ = roc_curve(y_test_encoded, y_prob_baseline)

# Plot AUC-ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_baseline, tpr_baseline, label=f'Baseline Model (AUC = {roc_auc_baseline:.4f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
plt.title('AUC-ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

The baseline model has an AUC of 0.6953.
